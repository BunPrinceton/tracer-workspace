```python
# Cave client, created Allen Institute, MICrRONS team (Seung lab, Baylor 
# College of Med, Harvard), is a Python client library that provides a 
# user-friendly, programmatic interface for scientists to query and analyze 
# the massive connectomics data stored in the CAVE system #

from caveclient import CAVEclient

#Creating a new function #

def get_all_stacks():
    """Get a list of all the currently-documented CAVE datastack names.
    
    Returns:
    stacks -- a list of all the current datastack names (list of str)"""
    
    # Gives previoiusly imported CAVEclient (an instance of a client) a 
    # function and shorthand name #
    # Client = lawyer #
    # "client" can be called anything you want #
    client = CAVEclient()

    # creates a new variable called "stacks". pulls list of all 
    # currently-documented datastacks #
    stacks = client.info.get_datastacks()
    
    # Return is the final part of the get_all_stacks function. It's a 
    # command that calls back the list (of strings) below #
    return stacks

get_all_stacks()

```

```python
from caveclient import CAVEclient

def get_stack_data(datastack):
    """Get all the metadata for a specific CAVE datastack.
    
    Argument:
    datastack -- the name of the datastack you want information for, e.g. 
    brain_and_nerve_cord (str)
    
    Returns:
    stack_info -- all the metadata on the requested datastack (dict)"""
    #the above code is defining (creating) a new function (procedure) called 'get_stack_data' that takes in the name of a brain collection (the argument being presented/input) in place of 'datastack' between the paranthesis 

    # Firstly, this is defining an a new instance of CAVEclient (The librarian) as client. So the conversation is something like this (though in a strange order in the code):
    # CAVEclient(....)     "Hi can I get a librarian to help me"
    # client =             "I'm going to call you 'client' from now on". 
    # datastack_name=     "Your sole job is to provide a dataset when asked"
    # =datastack           The   
    client = CAVEclient(datastack_name=datastack)

    # pulls datastact info dictionary using stack name #
    stack_info = client.info.get_datastack_info()

    return stack_info
```
